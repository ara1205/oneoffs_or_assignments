```python
class LinkedList:
    
    class Node:
        def __init__(self,val,next = None):
            self.val = val
            self.next = next
        
    def __init__(self):
        self.head = None
    def addHead(self,val):
        self.head = self.Node(val,self.head)
    def printList(self):
        node = self.head
        while node is not None:
            print(node.val,"--> ",end="")
            node = node.next
        print("*")
        
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
    #desc: counts how many elements are in a linked list
    #inputs: LinkedList (self cause its a function of a linked list mylist.count())
    #outputs: int (the count)
    def count(self):
        if self.head is None:
            print("0 elements")
            return
        else:
            node = self.head
            count = 0
            while node is not None:
                count = count + 1
                node = node.next
            print(count," elements")
        
        return count
    
    #desc: removes an element from a linked list and relinks it after
    #inputs: LinkedList, int(the location index of element starting a 0)
    #outputs: null
    def remove(self,loc):
        if self.head is not None:
            if loc <= self.count() - 1:
                if loc > 0:
                    node = self.head
                    pnode = None
                    count = 0
                    while count < loc:
                        pnode = node
                        node = node.next
                        count = count + 1

                    pnode.next = node.next
                    del node

                else:
                    self.head = self.head.next

                print("Successful Operation")
            else:
                print("out of bounds, range = 0 -",self.count() - 1)
        else:
            print("List not initialized")
        
        return
    #desc: adds an element from a linked list and relinks it after
    #inputs: LinkedList, int(the location index of the element you want to insert after, starting at 0), int (value of the element you are adding)
    #outputs: null
    def add(self,loc,val):
        if self.head is not None:
            if loc <= self.count() - 1:
                if loc > 0:
                    node = self.head
                    count = 0
                    while count < loc:
                        node = node.next
                        count = count + 1
                    newnode = self.Node(val,node.next)
                    node.next = newnode
                    print("Successful Operation")
                else:
                    self.addHead(val)


            else:
                print("out of bounds, range = 0 -",self.count() - 1)
        else:
            print("List not initialized")
        
        return
    #desc: adds all the values of a linked list into a pylist then prints it
    #inputs: LinkedList
    #outputs: null
    def pyPrint(self):
        if self.head is not None:
            
            list = []
            node = self.head
            while node is not None:
                list.append(node.val)
                node = node.next
            
            print(list)
            
        else:
            print("Empty List")
            
```

# Question 2

In the assignment it said that the first column will always be numbers but it didnt specify for the column labels. In the examples on the next page, the table has labels which are strings not numbers. The program was created with the assumption that the csv file will have labels in the first line.

```python
import csv

#desc: reads csv and turns into nested lists, also seperates label list
#inputs: str(name of csv file)
#outputs: nested lists (the data), list of str (the column labels list)
def datainput(filename):
    datafile = open(filename)
    dataRows = datafile.read().split("\n")
    datafile.close()
    data = []

    for row in dataRows:
        data.append(row.split(","))
    
    
    labels = data[0]
    #Del following lines if no labels~~
    data.pop(0)
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
    
    return(data,labels)

#desc: uses a Merge Sort to sort nested lists based on first element in each nested list
#inputs: nested lists (the data)
#outputs: nested lists (the data sorted by first element)
def listmergesort(data):
    
    
    if len(data) > 1:
        
        middle = len(data) // 2
        
        left = data[:middle]
        right = data[middle:]
        left = listmergesort(left)
        right = listmergesort(right)
        
        data = []
        
        while len(left) > 0 and len(right) > 0:
            
            #Extracts the first element of a list of lists, sadly 'left[0[0]]' isnt proper syntax
            leftnum = left[0]
            rightnum = right[0]
            
            if leftnum[0] < rightnum[0]:
                data.append(left[0])
                left.pop(0)
            else:
                data.append(right[0])
                right.pop(0)
        
        for x in left:
            data.append(x)
        
        for x in right:
            data.append(x)
            
    return data

#desc: opens and sorts a csv file by the first element per line, then saves it as new csv file
#inputs: str (filename of csv to open), str (filename of csv to save as)
#outputs: null
def sortCSV(f1,f2):
    
    data, labels = datainput(f1)
    sorteddata = listmergesort(data)
    
    #Del following lines if no labels~~
    sorteddata.insert(0,labels)
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
    
    datacopy = sorteddata.copy()
    print("Saved as csv file")
    
    with open(f2, "w", newline="") as f:
       writer = csv.writer(f)
       writer.writerows(datacopy)
    with open(f2, newline='') as csvfile:
     datacopy = csv.reader(csvfile, delimiter=' ')
    
    print(labels)
    print(data)
    print(sorteddata)
```

```python
#code to test which assumes the csv file is name 'test' and it saves the sorted data as 'output.csv'
filename = "test.csv"
sortCSV(filename,"output.csv")
```
